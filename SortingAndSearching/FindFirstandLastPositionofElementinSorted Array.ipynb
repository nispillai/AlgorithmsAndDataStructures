```python
class Solution:
    def findRange(self, nums, lo, hi, target, tRange):
        
        if hi < lo:
            return tRange
        mid = (lo + hi) // 2

        if nums[mid] == target:
            if tRange[0] == -1:
                tRange = [mid, mid]
            elif mid < tRange[0]:
                tRange = [mid, tRange[1]]
            elif mid > tRange[1]:
                tRange = [tRange[0], mid]
            if (mid - 1) >= 0 and nums[mid - 1] == target:
                tRange = self.findRange(nums, lo, mid - 1, target, tRange)
            if  (mid + 1) <len(nums) and nums[mid + 1] == target:
                tRange = self.findRange(nums, mid + 1, hi, target, tRange)
        elif tRange[0] == -1:
            if nums[mid] > target:
                tRange = self.findRange(nums, lo, mid - 1, target, tRange)
            else:
                tRange = self.findRange(nums, mid + 1, hi, target, tRange)
        else:
            if mid < tRange[0]:
                tRange = self.findRange(nums, mid + 1, hi, target, tRange)
            elif mid > tRange[1]:
                tRange = self.findRange(nums, lo, mid - 1, target, tRange)
            else:
                tRange = self.findRange(nums, lo, tRange[0] - 1, target, tRange)
                tRange = self.findRange(nums, tRange[1] + 1, hi, target, tRange)
                # I might need to add something here. 
        
        return tRange
        
    def searchRange(self, nums, target):
        lo = 0
        hi = len(nums) - 1
        tRange = [-1,-1]
        
        if hi < lo:
            return tRange
        return self.findRange(nums, lo, hi, target, tRange)
        
        
```

```python
class Solution1:
        
    def findBound(self, nums, target, isFound):
        lo = 0
        hi = len(nums) -1
        while lo <= hi:
            mid = (lo + hi) // 2
            if nums[mid] == target:
                if isFound:
                    if hi == mid or nums[mid + 1] != target:
                        return mid
                    lo = mid + 1
                else:
                    if lo == mid or nums[mid-1] != target:
                        return mid
                    hi = mid - 1
                    
            elif nums[mid] < target:
                lo = mid + 1
            else:
                hi = mid - 1
        return -1
        
    def searchRange(self, nums, target):
        lBound = self.findBound(nums, target, False)
        if lBound == -1:
            return [-1, -1]
        rBound = self.findBound(nums, target, True)
        return [lBound, rBound]
```
